```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
COVID19= pd.read_csv('esusLeitoOcupacao.csv', sep=",", encoding='UTF-8')
```

```python
COVID19.columns.values
```

```python
ESTADOS_COVID19=['dataNotificacao','ocupacaoConfirmadoCli',
       'ocupacaoConfirmadoUti','saidaConfirmadaObitos',
       'saidaConfirmadaAltas','estado']
```

```python
COVID = COVID19.filter(items=ESTADOS_COVID19)
```

```python
plt.rcParams['figure.figsize']=[50,40]
plt.style.context('fivethirtyeight')
plt.title('Evolução Clinica Por Milhão Dos Casos De COVID19 No Brasil Entre Abril, 2020 e Agosto, 2021', fontsize= 50, color='blue')
plt.xticks(fontsize=50, color='green')
plt.yticks(fontsize=50, color='blue')


plt.xlabel('Notificações 20 Mi',fontsize=50)
plt.ylabel('Evolução Dos Casos - 700 Mil - Não Acumulado',fontsize=50)

plt.hist(A, bins=range(100,210,5), alpha=1, stacked=False, color='blue')
plt.hist(B, bins=range(100,210,5), alpha=1, stacked=False, color='purple')

plt.legend(['Internação Clinica','Internação Uti','Alta','Obito'],loc='upper right', facecolor='Gray', fontsize=70)

plt.figure()
plt.savefig('internacao.png')

           
```

```python
plt.rcParams['figure.figsize']=[50,40]


plt.title('Mortes Por Casos De COVID19 No Brasil Entre Abril, 2020 e Agosto, 2021 Segundo O PORTAL BRASILEIRO DE DADOS ABERTOS', fontsize= 44, color='black')
plt.xticks(fontsize=50, color='orange')
plt.yticks(fontsize=50, color='red')

plt.xlabel('Evolução Dos Casos Acumulados - 400 MIL',fontsize=40)
plt.ylabel('Notificações - 20 Mi',fontsize=50)

plt.hist(D, bins=range(100,210,5), color='black',orientation="horizontal", cumulative=True)

plt.legend(['Obito'],loc='lower right', facecolor='Gray', fontsize=60)
```

```python
plt.rcParams['figure.figsize']=[70,50]


plt.title('Alts Apois Internação Na UTI Por Casos De COVID19 No Brasil Entre Abril, 2020 e Agosto, 2021 Segundo O PORTAL BRASILEIRO DE DADOS ABERTOS', fontsize= 50, color='black')
plt.xticks(fontsize=70, color='orange')
plt.yticks(fontsize=70, color='red')

plt.xlabel('Evolução Dos Casos Acumulados - 1 Mi',fontsize=50)
plt.ylabel('Notificações - 20 Mi',fontsize=50)

plt.hist(C, bins=range(100,210,5), alpha=0.8, stacked=False, color='green',orientation="horizontal",cumulative=True)


plt.legend(['Alta'],loc='lower right', facecolor='Gray', fontsize=80)
```
